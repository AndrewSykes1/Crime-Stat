```python
import csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.ticker import MaxNLocator
import numpy as np
```

```python
def csv_to_dict(file_path):
    data = []
    with open(file_path, mode='r') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            data.append(dict(row))
    return data
```

```python
def Plotter(data):
    date_list = []
    unrate = []

    for item in data:
        
        unrefined_date = item['DATE']
        year, month, trash = unrefined_date.split('-')
        date = round(int(year) + int(month)/12,3)
        date_list.append(date)

        unrate.append(float(item['UNRATE']))

    plt.figure(figsize=(40, 10))
    plt.plot(date_list, unrate)
    plt.xlabel('Year (Yr)')
    plt.ylabel('Unemployment Rate (%)')
    plt.title('Unemployment Rate vs Time')
    plt.gca().yaxis.set_major_locator(MaxNLocator(nbins=10))

    for i in range(1948,2025,4):
        plt.axvline(x=i, color='r', linestyle='--', label=f'x = {i}')
        plt.axvline(x=i-.4, color='b', linestyle='--', label=f'x = {i-.4}')
        plt.axvline(x=i+2, color='g', linestyle='--', label=f'x = {i+2}')

    return date_list, unrate

        
```

```python
path = r'C:\SZvsWorkspace\PersonalProjects\Crime_Analysis\UNRATE.csv'
data = csv_to_dict(path)
```

```python
print(Plotter(data))
```
