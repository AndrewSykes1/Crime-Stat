```python
import csv
import matplotlib.pyplot as plt
```

```python
def csv_to_dict(file_path):
    data = []
    with open(file_path, mode='r') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            data.append(dict(row))
    return data
```

```python
file_path = r'C:\SZvsWorkspace\PersonalProjects\Crime_Analysis\arrests_national_adults.csv'
crime_data = csv_to_dict(file_path)
print(f'Length of data: {len(crime_data)}')
print(crime_data[0])

crime_list = []
for item in crime_data:
    crime_name = item['offense_name']
    if crime_name not in crime_list:
        crime_list.append(crime_name)
    
print(crime_list)
```

```python
def Crime_over_time_grapher(crime_data, crime_name):
    
    races = ['white','black','asian_pacific_islander','american_indian']

    crime_dict = {}
    for year in range(1994,2017):
        crime_dict[year] = 0

    for item in crime_data:
        if item['offense_name'] == crime_name:
            
            total_crime = 0
            for name in races:
                total_crime += int(item[name])

            crime_dict[int(item['year'])] = total_crime
    
    year_list = []
    amount_list = []
    for key in crime_dict.keys():
        year_list.append(key)
        amount_list.append(crime_dict[key])

    for i in range(1996,2017,4):
        plt.axvline(x=i, color='r', linestyle='--', label=f'x = {i}')

    plt.xlabel('Year')
    plt.ylabel(f'Amount of {crime_name} Recorded')
    plt.title(f'{crime_name} vs Time')
    plt.plot(year_list, amount_list)
    plt.show()

    return crime_dict, year_list, amount_list
```

```python
for item in crime_list:
    Crime_over_time_grapher(crime_data, item)
```
