```python
import csv
import matplotlib.pyplot as plt
```

```python
def csv_to_dict(file_path):
    data = []
    with open(file_path, mode='r') as file:
        csv_reader = csv.DictReader(file)
        for row in csv_reader:
            data.append(dict(row))
    return data
```

```python
file_path = r'C:\Users\drews\Downloads\dataverse_files\Firearm_suicide_homicide_dataset.csv'
data = csv_to_dict(file_path)
```

```python
def State_gun_time_finder(state):

    state_data = []

    for i in range(len(data)):
        if data[i]['state'] == state:
            state_data.append(data[i])

    date_list = []
    pop_list = []

    for i in range(len(state_data)):
        date_list.append(int(state_data[i]['ï»¿year']))
        pop_list.append(float(state_data[i]['total_population']))
    
    return date_list, pop_list
```

```python
for state in state_list:
    date_list, pop_list = State_gun_time_finder(state)
    if pop_list[-1] > 1*10**7:

        plt.xlabel('Year')
        plt.ylabel("Guns Owned (10 Million)")
        plt.title('Gun Ownership vs Time')
        plt.plot(date_list, pop_list, label = state)
        plt.legend()
```

```python
date_list, pop_list = State_gun_time_finder('California')

plt.ylabel("Guns Owned (10 Million)")
plt.xlabel('Year')
plt.title('California Gun Ownership vs Time')
plt.plot(date_list, pop_list)
plt.show()
```

```python
pop_difference_average = (pop_list[-1] - pop_list[0] + pop_list[-2] - pop_list[1] + pop_list[-3] - pop_list[2]) / 3
year_difference = date_list[-1] - date_list[0]

pop_slope = pop_difference_average / year_difference
```
